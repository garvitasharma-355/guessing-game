```python
import random
computer_guess=random.randint(1,10)
user_guess=int(input("guess a number between 1 an 10="))
attempt=1
while user_guess!=computer_guess:
  if user_guess<computer_guess:
    print("guess higher")
    user_guess=int(input("guess again: "))
    attempt+=1
  else:
    print("guess lower")
    user_guess=int(input("guess again: "))
    attempt+=1
print("you guessed it right in",attempt,"attempts")
```
